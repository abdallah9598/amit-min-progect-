### Question1:
Write a python program to print the following pattern using loops.

<br>5 4 3 2 1 
<br>4 3 2 1 
<br>3 2 1 
<br>2 1 
<br>1


```python
#write you solution here
c=5
for i in range(c+1):
    for j in range(c - i,0, -1):
        print(j , end=' ')
    print('')   
```

### Question2:
Write a function that arrange string characters such that lowercase letters should come first. The function takes one string parameter that contains a combination of the lower and upper case letters and it returns the arranged string. <br>Call the function and print its returned value.

- Given:<br>
&emsp;&emsp;str1 = PyNaTive
<br><br>
- Expected Output: yaivePNT


```python
#write you solution here
str1="PyNaTive"
a_lower=[]
a_upper=[]
for i in str1:
    if i.islower():
        a_lower.append(i)
    else:
        a_upper.append(i)   

arr = ''.join(a_lower + a_upper)  
```
